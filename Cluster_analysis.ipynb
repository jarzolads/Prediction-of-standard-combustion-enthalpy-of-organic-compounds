```python
import pandas as pd
import scipy.stats as stats

data = pd.read_csv("/content/drive/MyDrive/Artículo Índices Topológicos JCIM 2025/Archivos csv/All_clusters.csv")
variables = ['Estrada', 'Wiener', 'Gutman', 'Enthalpy']

#Function to calculate confidence intervals
def confidence_intervals(series, confidence=0.95):
    mean = series.mean()
    sem = stats.sem(series, nan_policy='omit')  # Standard error of the mean
    interval = stats.t.interval(confidence, len(series.dropna()) - 1, loc=mean, scale=sem)
    return interval

#Function to calculate statistics and confidence intervals by group
def calculate_cluster_stats(group):
    stats = group.describe().transpose()
    ci_95 = group.apply(lambda x: confidence_intervals(x, 0.95))
    ci_99 = group.apply(lambda x: confidence_intervals(x, 0.99))

    stats["95% CI Lower"] = ci_95.iloc[0]
    stats["95% CI Upper"] = ci_95.iloc[1]
    stats["99% CI Lower"] = ci_99.iloc[0]
    stats["99% CI Upper"] = ci_99.iloc[1]

    return stats.reset_index()

#Group data by cluster and calculate statistics
clustered_data = data.groupby("Cluster")[variables]

#Calculate statistics for all clusters and combine them into a DataFrame
clustered_stats = pd.concat(
    [calculate_cluster_stats(clustered_data.get_group(cluster)).assign(Cluster=cluster)
     for cluster in clustered_data.groups],
    ignore_index=True
)

#Rename columns for clarity
clustered_stats.rename(columns={
    'index': 'Variable',
    'count': 'Count',
    'mean': 'Mean',
    'std': 'Std Dev',
    'min': 'Min',
    '25%': '25%',
    '50%': 'Median',
    '75%': '75%',
    'max': 'Max'
}, inplace=True)

clustered_stats
```

```python
data = pd.read_csv("/content/drive/MyDrive/Artículo Índices Topológicos JCIM 2025/Archivos csv/All_clusters.csv")

#Select relevant variables
variables = ['Estrada', 'Wiener', 'Gutman', 'Enthalpy']

#Function to calculate confidence intervals
def confidence_intervals(series, confidence=0.95):
    mean = series.mean()
    sem = stats.sem(series, nan_policy='omit')  # Error estándar de la media
    interval = stats.t.interval(confidence, len(series.dropna()) - 1, loc=mean, scale=sem)
    return interval

#Function to calculate statistics and confidence intervals by group
def calculate_cluster_stats(group):
    stats = group.describe().transpose()
    ci_95 = group.apply(lambda x: confidence_intervals(x, 0.95))
    ci_99 = group.apply(lambda x: confidence_intervals(x, 0.99))

    stats["95% CI Lower"] = ci_95.iloc[0]
    stats["95% CI Upper"] = ci_95.iloc[1]
    stats["99% CI Lower"] = ci_99.iloc[0]
    stats["99% CI Upper"] = ci_99.iloc[1]

    return stats.reset_index()

#Group data by cluster and calculate statistics
clustered_data = data.groupby("Cluster")[variables]

clustered_stats = pd.concat(
    [calculate_cluster_stats(clustered_data.get_group(cluster)).assign(Cluster=cluster)
     for cluster in clustered_data.groups],
    ignore_index=True
)

#Rename columns for clarity
clustered_stats.rename(columns={
    'index': 'Variable',
    'count': 'Count',
    'mean': 'Mean',
    'std': 'Std Dev',
    'min': 'Min',
    '25%': '25%',
    '50%': 'Median',
    '75%': '75%',
    'max': 'Max'
}, inplace=True)

#Code for latex
latex_rows = ""
for cluster, group in clustered_stats.groupby("Cluster"):
    first_row = True
    for _, row in group.iterrows():
        cluster_text = str(cluster) if first_row else ""
        latex_rows += (
            f"{cluster_text} & {row['Variable']} & {int(row['Count'])} & {row['Mean']:.2f} & {row['Std Dev']:.2f} & "
            f"{row['Min']:.2f} & {row['25%']:.2f} & {row['Median']:.2f} & {row['75%']:.2f} & {row['Max']:.2f} & "
            f"{row['95% CI Lower']:.2f} & {row['95% CI Upper']:.2f} & {row['99% CI Lower']:.2f} & {row['99% CI Upper']:.2f} \\\\\n"
        )
        first_row = False

latex_table = f"""
\\begin{{sidewaystable}}
  \\caption{{Descriptive statistics and confidence intervals (95\\% and 99\\%) for each variable across clusters.}}
  \\label{{tbl:cluster_stats}}
  \\centering
  \\begin{{tabular}}{{llllllllllllll}}
    \\hline
    Cluster & Variable & Count & Mean & Std Dev & Min & 25\\% & Median & 75\\% & Max & 95\\% CI Lower & 95\\% CI Upper & 99\\% CI Lower & 99\\% CI Upper \\\\
    \\hline
    {latex_rows}
    \\hline
  \\end{{tabular}}
\\end{{sidewaystable}}
"""

print(latex_table)
```

```python
file_descriptors = "/content/drive/MyDrive/Artículo Índices Topológicos JCIM 2025/Archivos csv/All_molecularDescriptors.csv"
file_clusters = "/content/drive/MyDrive/Artículo Índices Topológicos JCIM 2025/Archivos csv/All_clusters.csv"

data_descriptors = pd.read_csv(file_descriptors)
data_clusters = pd.read_csv(file_clusters)


ClusterAna=pd.concat([data_clusters,data_descriptors.drop(["Name","Entalpia"], axis=1)],axis=1)
#ClusterAna.to_csv("ClusterAna.csv")
```

```python
#Filter the relevant descriptors
descriptors = ['Chi0n', 'Chi1n', 'NumValenceElectrons', 'SlogP_VSA5',
               'Kappa1', 'Chi2n', 'SMR_VSA5', 'Chi0', 'Chi1v', 'HeavyAtomCount']

cluster_column = 'Cluster'

#Select descriptors and the cluster column
filtered_data = ClusterAna[descriptors + [cluster_column]]

#Perform descriptive analysis and confidence interval calculations
results = {}

#Loop through each cluster
for cluster, group_data in filtered_data.groupby(cluster_column):
    stats = group_data.describe().transpose()

    #Confidence intervals at 95% and 99%
    confidence_95 = 1.96 * (stats['std'] / (stats['count'] ** 0.5))
    confidence_99 = 2.58 * (stats['std'] / (stats['count'] ** 0.5))

    stats['95% CI Lower'] = stats['mean'] - confidence_95
    stats['95% CI Upper'] = stats['mean'] + confidence_95
    stats['99% CI Lower'] = stats['mean'] - confidence_99
    stats['99% CI Upper'] = stats['mean'] + confidence_99

    results[cluster] = stats[['mean', 'std', 'min', '25%', '50%', '75%', 'max',
                               '95% CI Lower', '95% CI Upper',
                               '99% CI Lower', '99% CI Upper']]

#Combine results into a single DataFrame for display
final_results = pd.concat(results, keys=results.keys(), names=['Cluster', 'Descriptor'])

pd.set_option('display.float_format', '{:.4f}'.format)
pd.set_option('display.max_rows', None)
final_results
```
